```python
from factor_momentum import FACTORS

from sf_quant.data.assets import load_assets
from sf_quant.data.factors import load_factors

import polars as pl
import datetime as dt

import pandas as pd
import statsmodels.api as sm
```

```python
start = dt.date(2000,1,1)
end = dt.date(2024,1,1)


factor_returns = load_factors(start=start, end=end, factors=FACTORS)
factor_returns
```

```python
asset_returns = (load_assets(
    start=start, end=end, in_universe=True, columns=['date', 'barrid', 'return']
    )
    .pivot(
        on='barrid', index='date', values='return'
    )
    .sort('date')
    )
asset_returns
```

```python
def rolling_regression(y, X, window=90):
    """
    Returns a DataFrame of rolling betas (including alpha as 'const')
    for a single asset series y on factor matrix X.
    """
    betas = []
    idxs = []

    for i in range(window, len(y) + 1):
        y_win = y.iloc[i-window:i]
        # Align the indices safely
        X_win = X.loc[X.index.intersection(y_win.index)]

        # Intersect indices again just to be absolutely sure
        common_idx = y_win.index.intersection(X_win.index)
        y_win = y_win.loc[common_idx]
        X_win = X_win.loc[common_idx]

        # Skip window if too few valid obs
        if len(y_win) < window // 2:
            continue

        model = sm.OLS(y_win.values, sm.add_constant(X_win.values)).fit()
        betas.append(model.params)
        idxs.append(y.index[i-1])

    return pd.DataFrame(betas, index=idxs, columns=['const'] + list(X.columns))
```

```python
factor_pd = factor_returns.to_pandas().set_index('date')
asset_pd = asset_returns.to_pandas().set_index('date')
```

```python
print(asset_pd.index.equals(factor_pd.index))  # should print True
```

```python
window = 90

X = factor_pd.copy()
X = sm.add_constant(X)

rolling_betas = {}

for i, asset in enumerate(asset_pd.columns):
    y = asset_pd[asset].dropna()
    rolling_betas[asset] = rolling_regression(y, factor_pd, window)
    print(f'{i}th asset Finished')

# Combine into one MultiIndex DataFrame
combined = pd.concat(rolling_betas, axis=1)

combined
```
